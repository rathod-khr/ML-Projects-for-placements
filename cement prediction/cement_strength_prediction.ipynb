```python
# creating feature and label

X = data.iloc[:, :-1]
y = data.iloc[:, -1]
X
```

```python
X.var()
```

```python
# normalizing features
# let's add 1 to each value in everycolumn so that we don't get exception while calculating the log value of 0

for column in X.columns:
    X[column] += 1
    X[column] = np.log(X[column])
```

```python
# scaling data

from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
X_train
```

```python
from sklearn.linear_model import LinearRegression,Ridge,Lasso
from sklearn.ensemble import RandomForestRegressor,GradientBoostingRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import GridSearchCV
```

```python
model1=LinearRegression()
model2=Ridge()
model3=Lasso()
model4=RandomForestRegressor()
model5=GradientBoostingRegressor()
model6=DecisionTreeRegressor()
```

```python
model1.fit(X_train,y_train)
model2.fit(X_train,y_train)
model3.fit(X_train,y_train)
model4.fit(X_train,y_train)
model5.fit(X_train,y_train)
model6.fit(X_train,y_train)
```

```python
#Training Accuracy of different Models

print(" LR training Accuracy: ", model1.score(X_train,y_train))
print(" Ridge training Accuracy: ", model2.score(X_train,y_train))
print(" Lasso training Accuracy: ", model3.score(X_train,y_train))
print(" Random Forest training Accuracy: ", model4.score(X_train,y_train))
print(" Gradiant Boosting Training Accuracy: ", model5.score(X_train,y_train))
print(" Decision Tree training Accuracy: ", model6.score(X_train,y_train))
```

```python
#Testing Accuracy on test data of different Models

print(" LR Test Accuracy: ", model1.score(X_test,y_test))
print(" Ridge Test Accuracy: ", model2.score(X_test,y_test))
print(" Lasso Test Accuracy: ", model3.score(X_test,y_test))
print(" Random Forest Test Accuracy: ", model4.score(X_test,y_test))
print(" Gradiant Boosting Test Accuracy: ", model5.score(X_test,y_test))
print(" Decision Tree test Accuracy: ", model6.score(X_test,y_test))
```

```python
models = pd.DataFrame({
    'Model' : ['Linear Regression', 'Ridge Regression','Lasso Regression', 'Random Forest', 'Gradient Boost','Decision Tree'],
    'Score' : [model1.score(X_test,y_test), model2.score(X_test,y_test),model3.score(X_test,y_test),model4.score(X_test,y_test),model5.score(X_test,y_test),model6.score(X_test,y_test)]
})

models.sort_values(by = 'Score', ascending = False)
```

```python
#checking another accuracy parameter i.e MSE


y_pred1=model1.predict(X_test)
y_pred2=model2.predict(X_test)
y_pred3=model3.predict(X_test)
y_pred4=model4.predict(X_test)
y_pred5=model5.predict(X_test)
y_pred6=model6.predict(X_test)
```

```python
mse1 = mean_squared_error(y_test, y_pred1)
mse2 = mean_squared_error(y_test, y_pred2)
mse3 = mean_squared_error(y_test, y_pred3)
mse4 = mean_squared_error(y_test, y_pred4)
mse5 = mean_squared_error(y_test, y_pred5)
mse6 = mean_squared_error(y_test, y_pred6)
```

```python
print(" LR MSE Accuracy: ", mse1)
print(" Ridge MSE Accuracy: ",  mse2)
print(" Lasso  MSE Accuracy: ",  mse3)
print(" Random Forest MSE Accuracy: ", mse4)
print(" Gradiant Boosting MSE Accuracy: ", mse5)
print(" Decision Tree MSE Accuracy: ", mse6)
```

```python
models = pd.DataFrame({
    'Model' : ['Linear Regression', 'Ridge Regression','Lasso Regression', 'Random Forest', 'Gradient Boost','Decision Tree'],
    'MSE_Score' : [mse1, mse2, mse3, mse4, mse5,mse6]
})

models.sort_values(by = 'MSE_Score', ascending = False)
```
